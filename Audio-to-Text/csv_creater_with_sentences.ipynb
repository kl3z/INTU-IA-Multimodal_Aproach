```python
from docx import Document
import re
import whisper
import pandas as pd
import ffmpeg
```

```python
caminho_docx = r"C:\Users\garci\OneDrive\Área de Trabalho\Roteiro de Entrevista Simulad1.docx"
documento = Document(caminho_docx)
for paragrafo in documento.paragraphs:
    print(paragrafo.text)
```

```python
texto = "\n".join([p.text.strip() for p in documento.paragraphs if p.text.strip() != ""])
paragrafos = re.findall(r'(.+?)(?:\n|$)', texto, re.DOTALL)
paragrafos = [p.strip() for p in paragrafos if p.strip() != ""]
df = pd.DataFrame(paragrafos, columns=["frase"])
df.to_csv("interrogatorio_respostas.csv", index=False)
```
